```python
import requests
import json
import time
import csv
```

```python
url = 'https://www.instagram.com/graphql/query'

short_code = input('Masukan Kode : ')
#short_code = 'Cdc46gopgm3'

headers = {'cookie': 'sessionid=1914542516%3AZV8OHfmEyKaOfo%3A21%3AAYeStETzDsMIkrr8CKgJis_ajQJ1bDr5U63w7yaaPg'}

end_cursor = ''
count = 0
counter_file = 1
jumlah_per_file = 1000

writer = csv.writer(open('hasil-komentar/{} {}.csv'.format(short_code, counter_file), 'w', newline=''))
head = ['Username', 'Text']
writer.writerow(head)

while 1:
    var = {"shortcode": short_code, "first":50, "after": end_cursor}
    params = {
        'query_hash': '69cba40317214236af40e7efa697781d',
        'variables': json.dumps(var)
    }
    
    r = requests.get(url, headers=headers, params=params).json()
    
    try:
        users = r['data']['shortcode_media']['edge_media_to_parent_comment']['edges']
    except:
        print('wait 60 seconds')
        time.sleep(60)
        continue
        
    for user in users:
        if count % jumlah_per_file == 0 and count != 0:
            counter_file += 1
            writer = csv.writer(open('hasil-komentar/{} {}.csv'.format(short_code, counter_file), 'w', newline=''))
            head = ['Username','Text']
            writer.writerow(head)
        username = user['node']['owner']['username']
        text = user['node']['text']
        writer = csv.writer(open('hasil-komentar/{} {}.csv'.format(short_code, counter_file), 'a', newline='', encoding='utf-8'))
        data = [username, text]
        writer.writerow(data)
        count += 1
        print(count, username, text)
        
    end_cursor = r['data']['shortcode_media']['edge_media_to_parent_comment']['page_info']['end_cursor']
    hash_next_page = r['data']['shortcode_media']['edge_media_to_parent_comment']['page_info']['has_next_page']
    if hash_next_page == False:
        break
        
    time.sleep(10)
```
